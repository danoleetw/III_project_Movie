```python
# -*- coding: utf-8 -*-
"""
Created on 2016/11/13
ver 1.0.0 : 將觸電、豆瓣及開眼三個document裡的資訊合併成一個
@author: dano
"""
import os
import csv
import pymongo
from pymongo import MongoClient
from hanziconv import HanziConv

#讀取三合一大表
final_csv = r'E:\Final.csv'
with open(final_csv, 'r') as f:
    reader = csv.reader(f)
    list_3in1 = [row for row in reader]
    
#建立连接
client = MongoClient('localhost', 27017)
db_name = 'final' #数据库名
db = client[db_name]
collect_a = db.atmovie
collect_d = db.douban
collect_t = db.truemovie
collect_f = db.final
#去抬頭
list_3in1.remove(list_3in1[0])
#給個number列印用
num = 1
for row in list_3in1:
    print num
    douban_id = str(row[0])
    truemovie_url = str(row[1])
    atmovie_id = str(row[2])
    
    #抓三個document出來
    document_a = collect_a.find_one({'atmovie_id':atmovie_id})
    document_d = collect_d.find_one({'movie_id':douban_id})
    document_t = collect_t.find_one({'url':truemovie_url})
    
    #刪除原document裡的_id
    del document_a['_id']
    del document_d['_id']
    del document_t['_id']
    
    #開眼title_t去\n\t\r，並改title_t為title_at、改IMDB為IMDB_at、改rating為rating_at，刪除沒有使用的title_e、title_c、truemovie_url及type
    document_a['title_t'] = document_a['title_t'].replace('\n','').replace('\t','').replace('\r','')
    document_a.update({'title_at': document_a['title_t']})
    del document_a['title_t']
    document_a.update({'IMDB_at': document_a['imdb']})
    del document_a['imdb']
    document_a.update({'rating_at': document_a['rating']})
    del document_a['rating']
    del document_a['title_e']
    del document_a['title_c']
    del document_a['truemovie_url']
    del document_a['type']
    
    #觸電的title_t改為title_true、type改為type_true
    document_t.update({'title_true': document_t['title_t']})
    del document_t['title_t']
    document_t.update({'type_true': document_t['type']})
    del document_t['type']
    
    #豆瓣文字簡轉繁、改movie_id為douban_id、改IMDB為IMDB_dou、改rating為rating_dou、type改為type_dou
    document_d['title1']= HanziConv.toTraditional(document_d['title1'])
    if document_d['title2']:
        document_d['title2']= HanziConv.toTraditional(document_d['title1'])
    if document_d['type']:
        document_d['type']= HanziConv.toTraditional(document_d['type'])
    document_d.update({'douban_id': document_d['movie_id']})
    del document_d['movie_id']
    document_d.update({'IMDB_dou': document_d['IMDB']})
    del document_d['IMDB']
    document_d.update({'rating_dou': document_d['rating']})
    del document_d['rating']
    if document_d['type']:
        document_d.update({'type_dou': document_d['type']})
        del document_d['type']
   
    #寫入大document
    document_a.update(document_t)
    document_a.update(document_d)
    collect_f.insert_one(document_a)
    
    print document_d['douban_id']
    num += 1
print "====================完成啦================"
```
